```python
def get_prod_dir(crew_list):
  try:
    producer_list=[]
    director_list=[]
    for elem in crew_list:
      if elem['job']=='Producer':
        producer_list.append(elem['name'])
      if elem['job']=='Director':
        
        director_list.append(elem['name'])
    return producer_list,director_list
  except:
    return [],[]
```

```python
get_prod_dir(no_null_crew.loc[1,'crew'])
```

```python
imdb_df[['Producer','Director']]=imdb_df.apply(lambda row:get_prod_dir(row['crew']),axis=1,result_type="expand")
```

## Deep dive into the movies, genres and profits corresponding to this actor.

```python
profit1=[]
profit2=[]
movie1=[]
movie2=[]
for index,row in not_null_cast.iterrows():
  for iter in not_null_cast.loc[index,'cast']:
    actor=iter['name']
    if 'Samuel L. Jackson' in actor:
      profit1.append(not_null_cast.loc[index,'profit'])
      movie1.append(not_null_cast.loc[index,'original_title'])

    if 'Robert De Niro'in actor:
      profit2.append(not_null_cast.loc[index,'profit'])
      movie2.append(not_null_cast.loc[index,'original_title'])
```

```python
#we will get the genre of samual and robert movies

genre_s=[]
a=[]
for i in  range(len(movie1)):
  for g in not_null_cast.loc[i,'genres']:
    a.append(g['name'])
  genre_s.append(a)
a=[]



```

```python
genre_r=[]
b=[]
for i in range(len(movie2)):
  for g in not_null_cast.loc[i,'genres']:
    b.append(g['name'])
  genre_r.append(b)
b=[]
```

```python
gen_s=np.array(genre_s)
gen_r=np.array(genre_r)
```

```python
#creating sub dataframe for robert
mov1=pd.DataFrame(movie1,columns=['movie name'])
prof1=pd.DataFrame(profit1,columns=['movie profit'])
gens=pd.DataFrame(gen_s.flatten(),columns=['movie genre'])
```

```python
movie_by_robert=pd.concat([mov1,prof1,gens],axis=1)
```

```python
movie_by_robert.sort_values(by='movie profit',ascending=False).head()
```

```python
#creating sub dataframe for samual
mov2=pd.DataFrame(movie2,columns=['movie name'])
prof2=pd.DataFrame(profit2,columns=['movie profit'])
genr=pd.DataFrame(gen_r.flatten(),columns=['movie genre'])
```

```python
movie_by_samual=pd.concat([mov2,prof2,genr],axis=1)
```
